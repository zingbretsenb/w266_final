```python
import pandas as pd
import numpy as np
```

```python
def question_matrix(f):
    q_list = []
    with open(f) as f:
        for line in f:
            line = line.strip()
            if line == 'Incorrect :-(':
                q_list.append(0) 
            elif line == 'Correct!':
                q_list.append(1)
            elif line == 'Unknown words!':
                q_list.append('U')
    return q_list
```

```python
df = pd.read_csv('errors.csv')
df.drop(df.columns[0], axis=1, inplace=True)

file_list = ["glove27B_twitter_200_e.txt", "glove27B_twitter_200_c.txt", "glove42b_300_e.txt", "glove42b_300_c.txt","glove840b_300_e.txt", "glove840b_300_c.txt"]


for f in file_list:
    df[f] = pd.Series(question_matrix(f))

numerical = df.columns
```

```python
df.replace('U', np.NaN, inplace=True)
df = df.apply(pd.to_numeric)
df.sum()
```

```python
mdf = pd.read_csv('pos_and_src.txt')
mdf.columns = [col.strip() for col in mdf.columns]
for col in mdf.columns:
    mdf[col] = mdf[col].str.strip()
```

```python
mdf['same_POS'] = mdf["POS1"] == mdf["POS2"]
categorical = mdf.columns
mdf.head()
```
